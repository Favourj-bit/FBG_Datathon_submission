```python
# Convert 'Transaction Date and Time' to datetime
df['Transaction Date and Time'] = pd.to_datetime(df['Transaction Date and Time'])

# Extract date and time features

df['Transaction Day'] = df['Transaction Date and Time'].dt.day
df['Transaction Hour'] = df['Transaction Date and Time'].dt.hour
df['Transaction Day of Week'] = df['Transaction Date and Time'].dt.dayofweek
```

```python
df = df.drop(columns=['Transaction Date and Time', 'Merchant ID'],
                       axis=1)
```

```python
obj_ = df.select_dtypes(include='object')
```

```python
obj_.head()
```

```python
obj_.nunique()
```

```python
new_data = df.drop(columns=['Device Type','Browser Type','Operating System','User Occupation',
                                 'Transaction Purpose','Transaction Time of Day','User Occupation','User Gender'])
```

```python
obj_col = new_data.select_dtypes(include='object').columns
```

```python
obj_col = obj_col.drop('IP Address')
```

```python
obj_col = obj_col.drop("User's Email Domain")
```

```python
obj_col
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

# Apply LabelEncoder to categorical columns
for column in obj_col:
    new_data[column] = le.fit_transform(new_data[column])
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
# Count unique IP addresses per user (assuming no direct "User" column)
ip_counts = new_data.groupby('IP Address')['IP Address'].transform('count')

# Add the counts as a new column
new_data['Unique IP Count'] = ip_counts
```

```python
new_data['Email Domain'] = new_data["User's Email Domain"].str.split('.').str[0]
```

```python
category_counts = new_data['Email Domain'].value_counts()
new_data['Email Domain Frequency'] = new_data['Email Domain'].map(category_counts)
```

```python
new_data.describe()
```

```python
new_data.head()
```

```python
X = new_data.drop(columns=['IP Address', "User's Email Domain",'Email Domain','Fraudulent Flag'], axis=1)
```

```python
y = new_data['Fraudulent Flag']
```

```python
X.head()
```

```python
y.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.dtypes
```

```python
# Create a Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
rf.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = rf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Print a classification report for more detailed metrics
print(classification_report(y_test, y_pred))
```
